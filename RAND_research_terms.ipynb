```python
props = "Frequency, Cross-discipline, multi-discipline, longitudinal, institution comparability, formative, comprehensive, flexible, assessment scalability, initial burden, participant burden, central burden, needed participant expertise, needed central expertise, extent of implementation, central ownership, free from judgement, transparency, quantitative".split(", ")  
nprop = len(props)
pi={}
for i in range(0,nprop):
    pi[props[i]]=i
```

```python
xloc = 0.2 + 0.0*np.random.random_sample(size=nprop)
yloc = np.random.random_sample(size=nprop)
```

```python

ax = plt.figure(figsize=(8,10))

minsize=6
maxsize=48
fontdict = {'size': 20, 'weight': 'bold', 'color':[0.4,0.5,1]}
fontdict2=fontdict.copy()
fontdict2['color']=[1.0,0.2,0.2]
pl=plt.plot([1],[1])
pl[0].set_visible(False)

for i in range(0,len(STAR)):
    fontdict['size'] = (maxsize-minsize)*(STAR[i]-1)/7.0 + minsize
    
    ax.text( xloc[i] ,yloc[i] ,props[i], fontdict=fontdict, ha='right')
    
    fontdict2['size'] = (maxsize-minsize)*(REF[i]-1)/7.0 + minsize
    
    ax.text( xloc[i]+0.01 ,yloc[i] ,props[i], fontdict=fontdict2, ha='left')

plt.axis('off');
```
