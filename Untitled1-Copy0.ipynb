```python
import numpy as np;
from math import *;
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import collections
from matplotlib.colors import colorConverter
```

```python
%matplotlib inline
```

```python

xangle = 0*(np.pi/180.0);
yangle = np.pi/2.0 - np.pi/3.0 - xangle;
d = 10;
math.cos(xangle)
z = 0;
l1 = np.array( [d*cos(xangle),  d*sin(xangle), z] )
l2 = np.array( [d*sin(yangle),  d*cos(yangle), z] )

Nx = 10;
Ny = 10;

latticeCoords = np.zeros([3*Nx,3*Ny,3]);

npolys = Nx*Ny;
nv = 3*npolys;
verts = np.zeros([nv,3]);

v0 = [0,0,0];
for i in range(0,Nx):
    v1 = v0+l1;    
    for j in range(0,Ny):
        p = i*Ny + j;
        v = 3*p;
        v2 = v0 + l2*[(1 - 2*(j%2)), 1, 0];
        verts[v] = v0;
        verts[v+1] = v1;
        verts[v+2] = v2;
        
    v0 = v0 + l1*[1, (1 - 2*(i%2)), 0];
        
        
#collections.PolyCollections( )
```

```python
fig = plt.figure(figsize=(12,12), dpi=72)
ax = fig.gca(projection='3d')

vv = verts[:,0:2].tolist();
col = collections.PolyCollection( [vv], closed='false' );
col.set_edgecolor([1,0,0]);
col.set_facecolor([0,0,1])
ax.set_xlim3d(0,100);
ax.set_xlabel('x')
ax.set_ylim3d(0,100);
ax.add_collection3d(col)
ax.view_init(50,-90)
```
