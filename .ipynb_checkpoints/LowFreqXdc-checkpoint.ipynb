```python
uopen = asin((diam/2.0)/rcm)
alpha = acos((0.2/fw90z))
alpha*180/pi
```

```python
#alpha = pi*60.0/180.0

R1=geom.getRotZYZarray(0, alpha/2, 0)
R2=geom.getRotZYZarray(0,-alpha/2, 0)

#r1 = np.array

def forward1(vec3s):
    return geom.translate3vecs(vec3s, [0, 0, -rcm]).dot(np.transpose(R1))
def forward2(vec3s):
    return geom.translate3vecs(vec3s, [0, 0, -rcm]).dot(np.transpose(R2))

u1 = forward1(uxyz)
u2 = forward2(uxyz)


nzp=400
axp=np.linspace(1,2*rcm,nzp)
axvec = np.transpose(np.array([np.zeros(nzp), np.zeros(nzp), axp]))

axvec1 = forward1(axvec)
axvec2 = forward2(axvec)

P1=transducers.calc_pressure_profile(k0, u1, uamp, axvec1)

P2=transducers.calc_pressure_profile(k0, u2, uamp, axvec2)


fig = plt.figure(figsize=(12,12), dpi=72)

plt.subplot(211)

ax = fig.gca(projection='3d')

p=ax.plot3D(u1[:,0], u1[:,1], u1[:,2], 'b.')
p=ax.plot3D(u2[:,0], u2[:,1], u2[:,2], 'g.')
p=ax.plot3D(axvec1[:,0], axvec1[:,1], axvec1[:,2], 'b--')
p=ax.plot3D(axvec2[:,0], axvec2[:,1], axvec2[:,2], 'g--')
ax.set_xlim3d( [-2*rcm,2*rcm]);
ax.set_ylim3d( [-2*rcm,2*rcm]);
ax.set_zlim3d( [-2*rcm,2*rcm]);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init( 0, 90 )

plt.subplot(212)
p=ax.plot(axp,P1, 'b')
p=ax.plot(axp,P2, 'g')
```
