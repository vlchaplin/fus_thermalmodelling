# Trajectory Def

```python
maxR_mm = 10
turnspace_mm=3
minR_mm = turnspace_mm / 2.0
z_mm = 140;

nturns = ceil( (maxR_mm - minR_mm)/turnspace_mm )

avgSpeed_mm_per_sec=0.6
focalpoint_dwell_seconds=6
wait=2

turn_radii_mm = minR_mm + turnspace_mm*np.arange(0,nturns)
num_sonications_per_turn = (np.round( (2*pi*turn_radii_mm/avgSpeed_mm_per_sec) / focalpoint_dwell_seconds ) );

num_sonications_total = np.sum(num_sonications_per_turn,dtype=int)
focalpoint_coords_mm = np.zeros([num_sonications_total, 3])
pass_relative_amplitudes = np.zeros([num_sonications_total, 256])

turnIdxStart=0;
for n in range(0,nturns):
    ns = num_sonications_per_turn[n]
    phis = np.arange(1,ns+1)*(2.0*pi/ns) + 0.5*(2.0*pi/ns)
    x = turn_radii_mm[n]*np.cos(phis)
    y = turn_radii_mm[n]*np.sin(phis)
    z = z_mm*np.ones(ns);
    
    i = turnIdxStart
    
    focalpoint_coords_mm[i:(i+ns),0] = x;
    focalpoint_coords_mm[i:(i+ns),1] = y;
    focalpoint_coords_mm[i:(i+ns),2] = z;
    
    turnIdxStart += ns
    
plt.plot( focalpoint_coords_mm[:,0], focalpoint_coords_mm[:,1], '*-' )
```

## Transducer amplitudes

#### First get power factor

```python
uxyz = sonalleve.get_sonalleve_xdc_vecs()
f0 = 1.2e6
c0 = 1540
k0 = 2.0*pi*(f0/c0)
rho=1000;

Ispta0 = 1.25e7;
N = uxyz.shape[0]
uamp0 = np.ones(N) / N
P0 = sonalleve.calc_pressure_field(k0, uxyz, uamp0, np.arange(0.125, 0.15, 200), [0.0], [0.14])
I0 = np.abs(P0)**2 / (2.0*rho*c0)

powerRenorm = (Ispta0/np.max(I0))
I0 *= powerRenorm

#update uamp0
uamp0 *= sqrt(powerRenorm)
```

```python
d=0.004;
M=3
ring,radius = geom.ring(d,M)
h=radius
pxyz=ring;


uamp1 = transducers.get_focused_element_vals(k0, uxyz, pxyz, np.ones([M]), L1renorm=sqrt(powerRenorm) )
#uamp1 = sqrt(powerRenorm)*( uamp1  ) / sum(abs(uamp1))
```

```python
pass_relative_amplitudes = np.zeros([num_sonications_total, N], dtype=complex)
for n in range(0,num_sonications_total,1):
    pass_relative_amplitudes[n] = transducers.get_focused_element_vals(k0, uxyz, pxyz + focalpoint_coords_mm[n]*1e-3, np.ones([M]), L1renorm=sqrt(powerRenorm) )
```

```python
dx=0.01
dy=dx
xp = 1e-2*np.arange(-2,2,dx)
yp=xp
zp = 1e-2*np.array([10,11, 12, 13, 14, 15])

Nx = len(xp)
Ny = len(yp)
Nz = len(zp)

gxp,gyp,gzp = np.meshgrid(xp, yp, zp, sparse=False, indexing='ij')

accumEnergy = np.zeros([Nx, Ny, Nz])
```

```python
h5name = "C:\\Users\\Vandiver\\Data\\simulations\\circ_nfoc\\EnergyIntVary_2.hdf5"
```

```python
f = h5py.File(h5name, "w")

dset= f.create_dataset("geom/trajectory", data=1e-3*focalpoint_coords_mm)
dset.attrs['wait'] = wait
dset.attrs['dwell'] = focalpoint_dwell_seconds
dset.attrs['speed'] = avgSpeed_mm_per_sec
dset.attrs['maxR_mm'] = maxR_mm
dset.attrs['turnspace_mm'] = turnspace_mm
dset.attrs['Ispta'] = Ispta0

f.flush()

dset = f.create_dataset("geom/gridx", data=gxp)
dset.attrs['xp']=xp
f.flush()

dset = f.create_dataset("geom/gridy", data=gyp)
dset.attrs['yp']=yp
f.flush()

dset = f.create_dataset("geom/gridz", data=gzp)
dset.attrs['zp']=zp
f.flush()



MList = [1,2,3,4,5,6]
dlist = np.arange(0.001, 0.006, 0.00025);

for M in MList:
        
    if M==1:
        dL=[0.0];
    else:
        dL=dlist;
        
    for di in range(0,len(dL)):
        d=dL[di]
        ring,radius = geom.ring(d,M)
        h=radius
        pxyz=ring;
        uamp1 = transducers.get_focused_element_vals(k0, uxyz, pxyz, np.ones([M]), L1renorm=sqrt(powerRenorm) )

        pass_relative_amplitudes = np.zeros([num_sonications_total, N], dtype=complex)
        for n in range(0,num_sonications_total,1):
            pass_relative_amplitudes[n] = transducers.get_focused_element_vals(k0, uxyz, pxyz + focalpoint_coords_mm[n]*1e-3, np.ones([M]), L1renorm=sqrt(powerRenorm) )

        accumEnergy[:] = 0

        for n in range(0,num_sonications_total,1):
            print(n, end='\r')
            pr=transducers.calc_pressure_field(k0, uxyz, pass_relative_amplitudes[n], xp,yp,zp)
            Iac = np.abs(pr)**2 / (2.0*rho*c0)
            accumEnergy += Iac*focalpoint_dwell_seconds*dx*dy
        
        maxAcEn= np.max(accumEnergy)
        
        dset = f.create_dataset("%d/%d" % (M,di+1), data=accumEnergy)
        dset.attrs['d']=d
        dset.attrs['pxyz']=pxyz
        f.flush()
        

f.close()
```
