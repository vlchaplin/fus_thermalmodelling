```python
alpha = pi*90.0/180.0

R1=geom.getRotZYZarray(0, alpha/2, 0)
R2=geom.getRotZYZarray(0,-alpha/2, 0)

#r1 = np.array

def forward1(vec3s):
    return (vec3s + [0, 0, -rcm]).dot(np.transpose(R1))
def forward2(vec3s):
    return (vec3s + [0, 0, -rcm]).dot(np.transpose(R2))

u1 = forward1(uxyz)
u2 = forward2(uxyz)


nzp=600
axp=np.linspace(1,2*rcm,nzp)
axvec = np.transpose(np.array([np.zeros(nzp), np.zeros(nzp), axp]))

axvec1 = forward1(axvec)
axvec2 = forward2(axvec)
axvecZ = axvec-[0,0,rcm]
axvecX = axvecZ.dot(geom.getRotZYZarray(0,pi/2, 0))

umerged = np.concatenate([uxyz,u1,u2])
umergedamp = np.concatenate([uamp,uamp,uamp])

P1=transducers.calc_pressure_profile(k0, umerged, umergedamp, axvec1)

P2=transducers.calc_pressure_profile(k0, umerged, umergedamp, axvec2)

Px=transducers.calc_pressure_profile(k0, umerged, umergedamp, axvecX)

Pz=transducers.calc_pressure_profile(k0, umerged, umergedamp, axvecZ)

I1 = np.abs(P1)**2
I2 = np.abs(P2)**2
Ix = np.abs(Px)**2
Iz = np.abs(Pz)**2


fig = plt.figure(figsize=(16,6), dpi=72)

ax = fig.add_subplot(121, projection='3d')

p=ax.plot3D(u1[:,0], u1[:,1], u1[:,2], 'b.')
p=ax.plot3D(u2[:,0], u2[:,1], u2[:,2], 'g.')
p=ax.plot3D(axvec1[:,0], axvec1[:,1], axvec1[:,2], 'b--')
p=ax.plot3D(axvec2[:,0], axvec2[:,1], axvec2[:,2], 'g--')
p=ax.plot3D(axvecZ[:,0], axvecZ[:,1], axvecZ[:,2], 'r--')
p=ax.plot3D(axvecX[:,0], axvecX[:,1], axvecX[:,2], 'y--')
ax.set_xlim3d( [-2*rcm,2*rcm]);
ax.set_ylim3d( [-2*rcm,2*rcm]);
ax.set_zlim3d( [-2*rcm,2*rcm]);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init( 0, -90 )

ax2 = fig.add_subplot(122)
ax2.plot(axp,I1, 'b')
ax2.plot(axp,I2, 'g')
ax2.plot(axp,Iz, 'r')
ax2.plot(axp,Ix, 'y')
ax2.set_ylabel('HIFU Intensity [arb.units]')
ax2.set_xlabel('path distance [cm]')
```
